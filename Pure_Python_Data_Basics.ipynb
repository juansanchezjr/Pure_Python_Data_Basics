```python
# Import dependencies
# Dependencies are software components, commonly referred to as libraries or modules, imported based on your needs
# Since we're importing a csv, we're leveraging the csv module and its reader object 
import csv
from csv import reader
```

```python
# This block of code imports our csv file and separates the header from the rest of the data for ease of us

opened_file = open('YEM_5.13.csv')
read_file = reader(opened_file)
all_data = list(read_file)
header = all_data[0]
working_data = all_data[1:] 
```

```python
# The type() function allows us to determine what data type we're working with
# In this case we can see that header is a list
# This helps us understand how to work with this data type
type(header)
```

```python
# Here we've used the print() function which prints our header list
# As a reminder, header is a variable we used to capture the column names of our csv data
print(header)
```

```python
# Above we see we have an 'event_type' column, in order to see the contents of this column we'll run a For Loop
# In this example x is what's known as an iterator and can be named anything, such as i, row, etc...
# x will iterate through every row of our working_data list (visualize an excel/csv and reviewing every single row)
# Events is a variable indicating we want x to iterate through every row of column 7 'event_type'(visualize the excel/csv)
# We ultimately print the events and see all of the contents of the 'event_type'

for x in working_data:
    events = x[7]
    print(events)
```

```python
# The For Loop above returns all of the contents, but is difficult to read based on the sheer number of events
# We'll reconfigure the For Loop to just pull the unique values without the duplicates by adding an "if statement"
# Of note, we're going to put the unique events in a "global variable" outside of the For Loop 
# A global variable indicates the variable can be used or "called" outside of the For Loop
# This "if statement" is a roundabout way of getting our unique values 
# The append() method adds items to our empty 'unique_events' list
# We ultimately print our unique_events and get the values from the 'event_type' column 

unique_events = []

for i in working_data:
    events = i[7]
    if events not in unique_events:
        unique_events.append(events)
print(unique_events)
```

```python
# In this block of code we're creating a new data set comprised of only 'Explosions/Remote Violence'
# We start by creating an empty list with the variable name 'explosions_remote_violence_list'
# We iterate through our 'working_data' list with 'row' as our iterator
# We provide an explicit value 'Explosions/Remove violence' in our 'if statement' 
# We then append() the results into our 'explosions_remote_violence_list' list
# We ultimately print and receive the contents of our new 'explosions_remote_violence_list' 
# Of note, this is now a list we can call, further filter down into and refine

explosions_remote_violence_list = []

for row in working_data: 
    value = row[7]
    if value == 'Explosions/Remote violence':
        explosions_remote_violence_list.append(row)
print(explosions_remote_violence_list)
```

```python
# We're now exporting our refined list to a csv 
# This new csv only contains rows with the 'event_type' of 'Explosions/Remote violence'
# Below, the 'ERV' in 'ERV.csv' can be named anything
# The values inside of writerow() and writerows() are our 'header' and 'explosions_remote_violence_list' lists

with open('ERV.csv', 'w') as f:
    write = csv.writer(f)  
    write.writerow(header)
    write.writerows(explosions_remote_violence_list)
```
